```python
def tau(N,i):
    assert i==1, "Not implemented"
    
    def time(r, p):
        w1 = r + N-1
        w2 = 2*r + N-2
        tau = w1 / (N-1 + r*p)
        tau += (N-1)^2 * (r*p*w2 + w1*(N-2+r*p)) / ((N-1)*(r*p)^2 + r*p + N-2)
        tau /= N
        
        return tau
    
    return time
```
